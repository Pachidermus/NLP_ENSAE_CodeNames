# Introduction

Games are increasingly used as fields to apply Natural Language Processing techniques. *Codenames* is a popular Czech board game which is based on similarity between words. Two teams compete, each team having one "codemaster" and one or several guessers. 25 words (codename tiles) are placed on the table. The role of the guessers is to find their teamâ€™s words, while avoiding the words belonging to the other team, to the assassin or to civilian squares. The guessers give their answers based on single-word clues given by the spymaster. 

In 2019, the *Codenames AI Competition* was organized. It consisted in confronting different types of codemasters and guessers bots to find the combination of players with the best victory rate. Therefore, using this competition as an inspiration, this project aims at creating a bot which can play this game, either as an guesser or as the codemaster.

## Description of the rules

Two teams compete : the blue team and the red team. 

25 word tiles are on the table. Each of them belong either to the blue team, to the red team, to the assassin or to the civilians. Only 1 word belongs to the assassin. 

Each team has 1 codemaster and 1 or several guessers. In our case, we will only simulate games teams of 2 players (1 codemaster and 1 guesser). The codemasters are the only one who know the *map* of the game, i.e. to which category belongs each word on the board.

1. The game starts with one team's codemaster trying to find a clue word for his teammate, that is related to his team's words but is unrelated enough to the other words. Then, he gives a 1-word clue to his teammate, as well as the number words related to the clue.
2. The guesser tries to find his team's word(s). If he succeeds once, he can try again until he finds a tile which does not belong to his team. If the assassin is found, the game is over and the team that has found the assassin loses. Else, it is the other team's turn to play.
3. The game ends when the assassin is called or when all the tiles of one team are uncovered.

## Guesser and Codemaster using Word2Vec or GloVe

Bots which are based on Word2Vec or GloVe embedding follow similar logical processes. They are all based on their respective embedding's vocabulary. Word2Vec and GloVe's pretrained vocabularies are cleaned by removing stopwords, punctuations, html tags and other tokens containing punctuation characters.

The guessers classes `GuesserW2V` and `GuesserGlove` are respectively based on Word2Vec pre-trained embedding and on GloVe pre-trained embedding. They both use their respective embedding's vocabulary to find the word(s) minimizing the cosine distance to the clue word. `GuesserW2V` and `GuesserGlove` can be customized using a **maximum distance parameter** : words which cosine distance to the clue word is higher than this maximum threshold will not be guessed. 

The codemaster classes `CodemasterW2V` and `CodemasterGlove` are respectively based on Word2Vec pre-trained embedding and on GloVe pre-trained embedding. They both use their respective embedding's vocabulary to find a clue word and a set of team words optimizing our custom distance function. The more a team will fall behind its enemy, the more words the codemaster of the former team will try to make his teammate guess. We also make sure that the codemaster will not give a clue with same lemma or stem than one of the words to be guessed: for example, the codemaster will avoid giving the clue "bananas" for the word "banana". *Lancaster stemmer* and *WordNet lemmatizer* are both used to ensure this. `CodemasterW2V` and `CodemasterGlove` can be customized using several parameters :
* The distance function can be customized by changing the **weights associated with team, enemy, civilian or assassin's words**.  For example, if we consider that the codemaster absolutely needs to avoid assassin's word, we will assign a high weight to it.
* A **greed parameter** allows to define whether the codemaster will adopt a risky behavior or not. The higher this parameter is, the more impatient the codemaster will be, trying to make his teammate guess a lot more words if his team is near to winning.
* We can also force the codemaster to make his teammate **guess a fixed number of words** during the whole game, though it was mainly used for test purposes.
* The **vocabulary size** can be changed. The codemaster's vocabulary will consist on the `vocab_size` most frequent words from its embedding's cleaned initial vocabulary.

## Guesser and Codemaster using Synsets

Bots which are based on WordNet's synsets follow a totally different logic from those which are based on a word embedding. The second codemaster developed with WordNet requires to define an external vocabulary to select clues. As such, a vocabulary `cm_wordlist` of more than 7,000 words is defined. We can print 5 random words from this list :

```python
with open("players/cm_wordlist.txt") as f:
    cm_wordlist = f.read().split('\n')
np.random.choice(cm_wordlist, 5).tolist()
```

The guesser class `GuesserWordNet` is based on WordNet's synsets. It uses the synsets of all possible words to find the one(s) which maximize the Wu-Palmer similarity with the clue word. `GuesserWordNet` can be customized using a **minimum similarity parameter** so that it only guesses words that present a good-enough similarity.

Two codemaster classes are defined : `CodemasterWordNet_HYP` and `CodemasterWordNet_WUP`. Like the previous codemaster classes, those classes exploit *Lancaster stemmer* and *WordNet lemmatizer* to ensure that clues will not have the same stem or lemma as a word on the board. However, the ways they generate clues are very distinct:
* The `CodemasterWordNet_HYP` class uses Wu-Palmer similarity in order to find the best combination of words to be guessed by his teammate. To select possible clues, it regroups most similar words by taking their lowest common **hypernym**, creating a sort of decision tree where each node is a common hypernym. Finally, it returns the lowest common hypernym for the combination of words with the highest similarity, while ensuring that this hypernym is sufficiently far from bad words. Then, it gives the hypernym as the clue. 
* The `CodemasterWordNet_WUP` class exclusively uses the **Wu-Palmer similarity** in order to find the best combination of words to be guessed by his teammate. To do so, this class computes Wu-Palmer similarities between words on the board, and words from `cm_wordlist` with their respective **synsets**. The clue given by the codemaster will be the one which maximizes the similarity, while ensuring that the clue is as unrelated as possible to bad words. This algorithm has been greatly inspired by [this article](https://ojs.aaai.org//index.php/AIIDE/article/view/5239). To make computations faster, we only use a random subset of `cm_wordlist` to find clues at each step. This also allows for this algorithm to return different answers for a given situation.  


For those two codemaster classes, a **greed parameter** allows to define how risky the codemaster will be.

## Overall Remarks about our Bots

While heuristics used for our bots differ quite strongly depending on their basis, guessers and codemasters tend to follow the same philosophy:
* Codemasters try to give valuable clues that connect to several words from their team while avoiding words from the enemy team and the assassin.
* Guessers order their guesses based on a similarity measured between the clue and available words. Most of the time, they try to guess the exact number of words indicated by the codemaster.

As a matter of fact, our algorithm all suffer from the same problems when compared to humans:
* They **don't remember previous plays** and act blindly to what happened earlier. This means that codemasters may end up giving the exact same clue twice in a row if the guessers didn't find its associated words during the first try. Also, guessers won't remember their mistakes and won't try to guess the words that they previously missed.
* Our bots **don't really pay attention to what the enemy team is doing** and only care about the number of enemy cards that have been revealed. More importantly, guessers from a team do not try to understand what the enemy clue might have meant for their guessers. 

Nevertheless, we believe that our bots can reach good enough performances while not adding such complexity to their behaviors. 

For more information, we encourage you to take a look at our bots source code, available in the `players` folder: 
* `guesser_cm_w2v` refers to Word2Vec-based algorithms.
* `guesser_cm_glove` refers to GloVe-based algorithms.
* `guesser_cm_wordnet` refers to WordNet-based algorithms.

## Benchmark Description

In order to assess our bots' performance, the following setup has been has been designed:
* Teams of bots, composed of a codemaster and a guesser, are defined. Seeds used to generate games are fixed.
* For each seed, each team starts by confronting itself in a "mirror match".
* For each seed, each team faces every other team twice, starting the same match as red and as blue in order to avoid potential biases due to unbalanced word grids (some games are very unbalanced since words are chosen randomly).

When all games are over, valuable statistics such as win rates and average number of turns are collected to measure performances. 

We are mostly interested in the **win rates** of each team against each other.  Nevertheless, we also attach a strong importance to the **average number of turns** taken by the algorithms to reach the end of a game, which is a good indicator of the overall greediness of our bots (especially when it is measured on games where the assassin has not been called). Finally, we monitor the **frequency of the assassins** in our games. Calling the assassin is something that occurs rather rarely when playing with humans, and we would like our algorithms to avoid calling the assassin as it less exciting for the other team to win in such a manner.

### Not Mixing Embeddings: *Word2Vec&Word2Vec vs GloVe&GloVe*

```python
Game2(CodemasterW2V, GuesserW2V, CodemasterGlove, GuesserGlove, seed=7, do_print=True,  
      cmr_kwargs={'model':model_w2v, 'vocab_size':4000, 'greed': 0.1, 'printed_dist':False}, 
      gr_kwargs={'model':model_w2v,'printed_dist':False, 'max_dist':2}, 
      cmb_kwargs={'model':model_glove300, 'red_team':False,'vocab_size':4000, 'greed': 0.1, 'printed_dist':False},  
      gb_kwargs={'model':model_glove300,'printed_dist':False},
      game_name="W2VvsGlove_7").run()
```

### Mixing Embeddings: *Word2Vec&GloVe vs GloVe&Word2Vec*

```python
Game2(CodemasterW2V, GuesserGlove, CodemasterGlove, GuesserW2V, seed=7, do_print=True,  
      cmr_kwargs={'model':model_w2v, 'vocab_size':4000, 'greed': 0.1, 'printed_dist':False}, 
      gr_kwargs={'model':model_glove300,'printed_dist':False, 'max_dist':2}, 
      cmb_kwargs={'model':model_glove300, 'red_team':False,'vocab_size':4000, 'greed': 0.1, 'printed_dist':False},  
      gb_kwargs={'model':model_w2v,'printed_dist':False},
      game_name="W2VGlovevsGloveW2V_7").run()
```

Generally speaking, in both games, **codemasters grow greedier as the game progresses**. They start by trying to make their teammates guess 2 words, and increase their expectations to 3 or 4 words in late-mid game (if enough words are remaining). As a consequence, first clues make more sense to humans than the last ones. For instance, the second clue given in both games was *'pakistan 2'* for *'china'* and '*india*', which are all countries located on the same continent. However, in the first game, one of the clue given by red team in late-mid game was *'trained'* for *'laser'*, *'horse'*, *'carrot'*. The relationship between the clue and this combination of words appears hard to understand for us. Since the AI codemasters get greedier as the game progresses, the distance between the clue given and the combination of words to be guessed also tends increase in average. Therefore, it also leads to more mistakes by the guesser. Indeed, in the first game, the first mistake happened late in the game (at turn 5), with guess *'pumkin'* for clue *'religious 4'.

It is interesting to see that even from the very first turn, **GloVe and Word2Vec bots do not make identical guesses** : in both game, the first clue was *'teacher 2'*. In both games, the guessers' first answer turn were identical. However, for this clue, the Word2Vec guesser gave *'flute'* (a correct word) as a second answer whereas GloVe guesser gave *'scientist'*, calling an enemy word. In total, 4 guesses mistakes were made in the first game whereas in the second game, 6 mistakes were made. A mistake was already made at the second turn in the second game, while no mistakes were made before turn 5 in the first game. Those elements suggest that Word2Vec codemaster works better with a Word2Vec guesser and the same observation can be stated for GloVe bots. Indeed, as Word2Vec and GloVe embeddings have distinct latent spaces, using bots with different embeddings in a team leads to more mistakes. 

Lastly, the first game lasted 9 turns and the winner was the red team, whereas the second game lasted 8 turns and the winner was the blue team. It seems that the duet (`CodemasterW2V`, `GuesserGlove`) performs worse than the duet (`CodemasterGlove, GuesserW2V`) in this specific game example.

# Conclusion and Perspectives

Using various NLP techniques, we managed to implement bots playing different types of players, which are codemaster and guesser. We compared the use of word embeddings such as Word2Vec and GloVe with methods based upon synsets. We exploited metrics such as the cosine distance or the Wu-Palmer similarity to build those bots. We also tried to take into account game environment to some extent, adding weights and greed parameters to our AI players. Our **final bots make guesses and clues which generally make sense** to us and they are able to play against - or with - humans. The benchmark results suggest that bots based on the use of hypernyms (first version of WordNet bots) are the worst performing ones, whereas the best ones are those based on GloVe embedding.

**Several improvements may be added** to our bots. First of all, as computing time would explode otherwise, we had to put a constraint on the the vocabulary size of the codemaster. A wider vocabulary could be used with proper optimizations of the current code. The second WordNet codemaster is especially time-consuming. For similar reasons, we could not execute a high number of games with different bots or different parameters values. We had to limit ourselves to only five different teams in our benchmark, which is good but does not allow to explore all possibilities. In addition, our bots do not take into account different aspects of game progress. As a consequence, our bots could not modulate their answers depending on the past guesses or clues, either from their own team or the enemy team. Adding this feature would allow reinforcing our bots and make their behavior more human-like. Finally, our implementation only allows teams of two players. To be able to run a game with several guessers in the same team, one solution may be to proceed to a majority poll system to decide for a team's final guesses.
